```python
# dividing the data into two parts input and output columns
x=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
#library used train test and split of the data
from sklearn.model_selection import train_test_split
# test_size is usesd to tell how much data should be used for training and test here 0.1 means that 10% data will be used for test and 90% for taining
train_test_split(x,y,test_size=0.1)
# the order in which the output is given
# x_train
# x_test
# y_train
# y_test
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
```

```python
x_train
# randomly rows are selceted for train data it is about 70%
```

```python
y_test
# randomly the rows are selceted for y_test
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_train
# the values will be inbetween -1 to 1
```

```python
x_test=scaler.transform(x_test)
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression()
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
#here we are checking the predictionr results by giving the test data
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x_train.values,y_train.values,clf=clf,legend=2)
```
